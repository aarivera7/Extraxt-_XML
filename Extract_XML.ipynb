```python
import pandas as pd
from xml.dom import minidom
from xml.etree.ElementTree import parse
```

```python
dataCR = pd.read_xml('data/TASS2019_country_CR_train.xml')
dataMX = pd.read_xml('data/TASS2019_country_MX_train.xml')
dataPE = pd.read_xml('data/TASS2019_country_PE_train.xml')
dataUY = pd.read_xml('data/TASS2019_country_UY_train.xml')
```

```python
documentCR = parse('data/TASS2019_country_CR_train.xml')
print(documentCR)
```

```python
documentMX = parse('data/TASS2019_country_MX_train.xml')
print(documentMX)
```

```python
documentPE = parse('data/TASS2019_country_PE_train.xml')
print(documentPE)
```

```python
documentUY = parse('data/TASS2019_country_UY_train.xml')
print(documentUY)
```

```python
# Esta función extrae en una lista los sentimientos del archivo xml
def extraerSentimiento(document):
    listaSentimiento=[]
    
    for item in document.iterfind('tweet'):
        listaSentimiento.append(item.find('sentiment').find("polarity").findtext('value'))
        
    return listaSentimiento
```

```python
sentimentCR = extraerSentimiento(documentCR)
print(sentimentCR)
print(len(sentimentCR))
```

```python
sentimentMX = extraerSentimiento(documentMX)
print(sentimentMX)
print(len(sentimentMX))
```

```python
sentimentPE = extraerSentimiento(documentPE)
print(sentimentPE)
print(len(sentimentPE))
```

```python
sentimentUY = extraerSentimiento(documentUY)
print(sentimentUY)
print(len(sentimentUY))
```
